```python
import scanpy as sc
import pandas as pd
```

```python
adata = sc.datasets.pbmc3k_processed()
```

```python
sc.pl.umap(adata, color="louvain")
```

```python
adata.obs["immune"] = "innate"
adata.obs.loc[
    adata.obs.louvain.isin(["CD4 T cells", "CD8 T cells", "B cells"]), "immune"
] = "adaptive"
```

```python
sc.pl.umap(adata, color="immune")
```

```python
pd.DataFrame(
    adata.X, index=adata.obs_names.values, columns=adata.var_names.values
).to_csv("data/pbmc_3k_proc.csv")
```
